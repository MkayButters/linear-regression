```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv("BankChurners.csv")
train.info()
```

```python
train = train.dropna()
```

```python
train["Customer_Age"] = train["Customer_Age"].apply(round)
train["Customer_Age"].head()
train["Months_on_book"] = train["Months_on_book"].apply(round)
train["Months_on_book"].head()
```

```python
X = train['Customer_Age'].values
y = train['Months_on_book'].values
```

```python
X=X.reshape(-1,1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y,train_size=0.8,test_size=0.2,random_state=100)
print(f"X_train shape {x_train.shape}")
print(f"y_train shape {y_train.shape}")
print(f"X_test shape {x_test.shape}")
print(f"y_test shape {y_test.shape}")
print(y_test)
```

```python
%matplotlib inline
plt.scatter(x_train,y_train,color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Bank Data')
plt.show()
```

```python
lm = LinearRegression()
lm.fit(x_train,y_train)
y_predict = lm.predict(x_test)
print(f"Train accuracy {round(lm.score(x_train,y_train)*100,2)} %")
print(f"Test accuracy {round(lm.score(x_test,y_test)*100,2)} %")
```

```python
plt.scatter(x_train,y_train,color='red')
plt.plot(x_test,y_predict)
plt.xlabel("Age Of Credit Card Holder")
plt.ylabel("Months With Bank")
plt.title("Credit Card Data")
plt.plot
```

I noticed definatly as the age goes up the months with the bank increases. It seems more maturity makes you stay with a bank longer.
